PREPROCESSING

```python
import os
import shutil
import pandas as pd
from pathlib import Path
```

```python
# 1. Configura i tuoi percorsi
FILTERED_DIR = Path('AudioSamplesFiltered')
PREPROCESSED_FOR_FILTERING_DIR = Path('AudioSamplesPreprocessed_ForFiltering')
PREPROCESSED_DIR = Path('AudioSamplesPreprocessed')
```

```python
if PREPROCESSED_DIR.exists():
    shutil.rmtree(PREPROCESSED_DIR)  # Rimuove la cartella e tutto il contenuto

PREPROCESSED_DIR.mkdir(parents=True) # La ricrea vuota
```

```python
filtered_files = set(os.listdir(FILTERED_DIR))

print(f"Trovati {len(filtered_files)} file.")
```

```python
copies_count = 0

for filename in os.listdir(PREPROCESSED_FOR_FILTERING_DIR):
    if filename in filtered_files:
        source_path = os.path.join(PREPROCESSED_FOR_FILTERING_DIR, filename)
        destination_path = os.path.join(PREPROCESSED_DIR, filename)
        
        # Copia il file da B a C
        shutil.copy2(source_path, destination_path)
        copies_count += 1

print(f"Operazione completata: {copies_count} file copiati in {PREPROCESSED_DIR}.")
```

FEATURE EXTRACTION

```python
CSV_INPUT_PATH = "audio_samples_features_for_filtering.csv"
CSV_OUTPUT_PATH = "audio_samples_features.csv"
```

```python
df = pd.read_csv(CSV_INPUT_PATH, index_col='filename')
```

```python
df_filtered = df.loc[df.index.intersection(filtered_files)]

df_filtered.to_csv(CSV_OUTPUT_PATH)

print(f"Salvate {len(df_filtered)} righe in {CSV_OUTPUT_PATH}")
```
